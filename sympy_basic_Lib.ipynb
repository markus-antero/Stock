```python
from sympy import (
    pprint,
    solve, 
    symbols, 
    init_printing, 
    Rational, 
    N, 
    expand, 
    factor,
    simplify,
    apart,
    together, 
    diff,
    integrate,
    Integral,
    Sum,
    Product,
    Eq,
    And
)
from sympy import plot_implicit
from sympy.plotting import plot

init_printing(use_latex='mathjax')
x, y, a, b, c = symbols("x y a b c")

expr = x**2 + y**2
expr2 = 1*x**2 + x*Rational(1,2) + 1
```

```python
eq1 = Eq(expr, 5)
p1 = plot_implicit(eq1,  (x, -3, 3), (y, -3, 3))
```

```python
plot((x**2, (x, -6, 6)), (x, (x, -5, 5)))
```

```python
print ("second order:")
pprint (expr2)
print ("circle:")
pprint (expr)

#substitute
print ("\n\nsubstitute - x = 1:")
pprint (expr.subs(x,1))
pprint (N(expr))
```

```python
#factor and expand & simplify
print ("\n\nfactor, expand, and simplify:\n")
pprint (expand(expr2))
pprint (factor(expr2))
pprint (simplify(expr2))

#apart & together
print ("\n\napart and together:\n")
pprint (apart(expr2))
pprint (together(expr2))
```

```python
from numpy import linspace
from sympy import lambdify, symbols
import matplotlib.pyplot as mpl

t = symbols('t')
x = 0.05*t + 0.2/((t - 5)**2 + 2)
lam_x = lambdify(t, x, modules=['numpy'])

x_vals = linspace(0, 10, 100)
y_vals = lam_x(x_vals)

mpl.plot(x_vals, y_vals)
mpl.ylabel("Speed")
mpl.show()
```
